```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Titanic-Dataset.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Age']=df['Age'].astype(int)
```

```python
df.isnull().sum()
```

```python
df['Cabin'].unique()
```

```python
df['Cabin']=df['Cabin'].fillna(df['Cabin'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['Embarked'].unique()
```

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Pclass'] = df['Pclass'].map({
    1: 'First Class',
    2: 'Second Class',
    3: 'Third Class'
})
```

```python
df.head()
```

```python
Age_Survived=df.groupby('Age')['Survived'].mean()
```

```python
df['Age_Survived']=df.groupby('Age')['Survived'].mean()
plt.figure(figsize=(12,6))
plt.plot(Age_Survived.index,Age_Survived.values, marker='o')
plt.xlabel("Age")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Age")
plt.grid(True)
plt.show()
```

```python
# Children (roughly ages 0–10) appear to have a relatively high survival rate.
# Most adult age groups (20–50) have moderate survival rates (around 20–50%).
# Older ages (60+) show large fluctuations because there were very few passengers in those age groups.
```

```python
sex_survival = df.groupby('Sex')['Survived'].mean()
```

```python
plt.figure(figsize=(6,6))
plt.pie(
    sex_survival.values,
    labels=sex_survival.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Survival Rate by Sex")
plt.show()
```

```python
# Female passengers had a significantly higher survival rate (around 80%) compared to male passengers (around 20%).
#This indicates that gender was one of the strongest factors affecting survival, likely due to the "women and children first" evacuation policy.
```

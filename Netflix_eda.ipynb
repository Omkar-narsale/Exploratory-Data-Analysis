```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('netflix_titles.csv')
```

```python
df['director']=df['director'].fillna('Unknown')
```

```python
df['cast']=df['cast'].fillna('Unknown')
```

```python
df['country']=df['country'].fillna('Unknown')
```

```python
df = df.dropna(subset=['date_added'])
```

```python
df['rating']=df['rating'].fillna(df['rating'].mode()[0])
```

```python
df['duration']=df['duration'].fillna(df['duration'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'].str.strip())
```

```python
df['date_added'].isnull().sum()
```

```python
df['Added_Year'] = df['date_added'].dt.year
df['Added_Month'] = df['date_added'].dt.month_name()
df['Added_Day'] = df['date_added'].dt.day
df['Added_Weekday'] = df['date_added'].dt.day_name()
```

```python
director = (
    df[df['director'] != 'Unknown']['director']
      .str.split(', ')
      .explode()
      .value_counts()
      .head(10)
)

plt.figure(figsize=(12,6))

ax = sns.barplot(
    x=director.values,
    y=director.index
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("🌍 Top 10 Directors Producing Netflix Content",
          fontsize=16,
          fontweight='bold')

plt.xlabel("Number of Titles")
plt.ylabel("Director")

plt.grid(axis='x', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python

# 📊 Observations
# Rajiv Chilaka has the highest number of Netflix titles (22).
# Jan Suter is second with 21 titles.
# Raúl Campos ranks third with 19 titles.
# The remaining directors have between 12 and 16 titles.
# The difference between the top-ranked directors is relatively small, indicating Netflix collaborates with multiple prolific directors rather than relying on just one.

# 💡 Business Insights
# Netflix works with a diverse set of directors instead of depending on a single filmmaker.
# Directors such as Rajiv Chilaka and Jan Suter appear to be frequent collaborators, suggesting strong production partnerships.
# The top 10 directors contribute multiple titles, helping maintain a consistent flow of content for the platform.

# ✅ Recommendation
# Netflix can continue strengthening relationships with its most productive directors while also exploring partnerships with emerging filmmakers to diversify its content portfolio.
```

```python
type_counts = df['type'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=90, colors=['#E50914', '#F4D03F'])
plt.title('Distribution of Netflix Content Types', fontsize=16, fontweight='bold')
plt.ylabel('') # Hide the default y-label if any
plt.show()
```

```python
#Movies dominate Netflix's content library, making up a significantly larger share than TV Shows.
#This indicates that Netflix has historically focused more on movie content than television series.
```

```python
year_type = (
    df.groupby(['Added_Year', 'type'])
      .size()
      .reset_index(name='Count')
)
plt.figure(figsize=(12,6))
sns.lineplot(
    data=year_type,
    x='Added_Year',
    y='Count',
    hue='type',
    marker='o'
)
plt.title("📈 Netflix Content Added Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.grid(alpha=0.3)
plt.show()
```

```python
# 📊 Observation
# The number of movies added to Netflix increased rapidly after 2016, reaching its highest point around 2019. TV Shows also grew steadily and reached their highest level around 2020.

# 💡 Business Insight
# This trend suggests that Netflix aggressively expanded its content library between 2016 and 2020, while also increasing its investment in TV Shows to meet changing viewer preferences.

# ✅ Recommendation
# Netflix should continue monitoring audience demand for Movies and TV Shows to maintain a balanced content library and optimize future content acquisition strategies.
```

```python
country = (
    df[df['country'] != 'Unknown']['country']
      .str.split(', ')
      .explode()
      .value_counts()
      .head(10)
)

plt.figure(figsize=(12,6))

ax = sns.barplot(
    x=country.values,
    y=country.index
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("🌍 Top 10 Countries Producing Netflix Content",
          fontsize=16,
          fontweight='bold')

plt.xlabel("Number of Titles")
plt.ylabel("Country")

plt.grid(axis='x', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Observation
# 🇺🇸 United States contributes the highest number of Netflix titles by a large margin.
# 🇮🇳 India is the second-largest contributor.
# 🇬🇧 United Kingdom ranks third (if "Unknown" is excluded).
# Japan and South Korea are among the top Asian content producers.
# Business Insights
# Netflix's catalog is heavily dominated by U.S. productions.
# India has become one of Netflix's most important international content markets.
# The strong presence of countries like Japan and South Korea reflects Netflix's investment in regional and global entertainment.
# Recommendation
# Continue investing in successful markets like India, South Korea, and Japan while also expanding content from underrepresented regions to diversify the catalog.
```

```python
# Remove Unknown countries
country_type = df[df['country'] != 'Unknown'].copy()

# Split multiple countries into separate rows
country_type = country_type.assign(
    country=country_type['country'].str.split(', ')
).explode('country')
```

```python
# Top 10 countries
top10 = country_type['country'].value_counts().head(10).index

plt.figure(figsize=(12,6))

ax = sns.countplot(
    data=country_type[country_type['country'].isin(top10)],
    y='country',
    hue='type',
    order=top10
)

# Add labels
for container in ax.containers:
    ax.bar_label(container)

plt.title(
    "🌍 Top 10 Countries by Movies and TV Shows on Netflix",
    fontsize=16,
    fontweight='bold'
)

plt.xlabel("Number of Titles")
plt.ylabel("Country")

plt.grid(axis='x', linestyle='--', alpha=0.3)

plt.legend(title='Content Type')

plt.tight_layout()
plt.show()
```

```python
# 📊 Possible Insights
# 🇺🇸 United States has the highest number of both Movies and TV Shows.
# 🇮🇳 India contributes mostly Movies compared to TV Shows.
# 🇰🇷 South Korea has a relatively higher proportion of TV Shows, reflecting Netflix's investment in Korean dramas.
# 🇯🇵 Japan contributes significantly to both Movies and TV Shows, including anime.
# 🇬🇧 United Kingdom has a balanced mix of Movies and TV Shows.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('googleplaystore.csv')
df.head()
```

```python
df_copy=df.copy()
```

```python
df_copy.drop(df_copy.index[10472])
```

```python

df_copy['Reviews'] = df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size']=df_copy['Size'].replace('M','000')
```

```python
df_copy['Size']=df_copy['Size'].replace('k','')
```

```python
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
```

```python
df_copy['Size'].unique()
df_copy['Size'] = df_copy['Size'].replace(',', '', regex=False)
df_copy['Size'] = df_copy['Size'].replace('+', '', regex=False)
df_copy['Size'] = df_copy['Size'].astype(float)
```

```python
df_copy.info()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price']=df_copy['Price'].str.replace('$','')
```

```python
df_copy['Installs'] = df_copy['Installs'].str.replace(',', '')
df_copy['Installs'] = df_copy['Installs'].str.replace('+', '')
```

```python
df_copy['Installs'] = df_copy['Installs'].astype(int)
```

```python
df_copy['Price']=df_copy['Price'].astype(float)
```

```python
df_copy.info()
```

```python
df_copy['Last Updated'].unique()
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Date']=df_copy['Last Updated'].dt.date
df_copy['Year']=df_copy['Last Updated'].dt.year
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Day']=df_copy['Last Updated'].dt.day
```

```python
df_copy.info()
```

```python
df_copy.to_csv('googleplaystore_cleaned.csv',index=False)
```

```python
df=pd.read_csv('googleplaystore_cleaned.csv')
df.head()

```

```python
df.shape
```

```python
df[df.duplicated('App')]
```

```python
df=df.drop_duplicates(subset=['App'],keep='first')
df.shape
```

```python
df['Category'].value_counts().plot.pie(figsize=(15,6),autopct='%1.1f')
plt.show()
```

```python
top_10_categories = df['Category'].value_counts().nlargest(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_categories.index, y=top_10_categories.values)
plt.title('Top 10 Categories of Apps')
plt.xlabel('Category')
plt.ylabel('Number of Apps')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df.head()
```

```python
df['Most_Installed'] = df.groupby('Category')['Installs'].transform('max')
```

```python
category_installs = df.groupby('Category')['Installs'].sum().reset_index()

plt.figure(figsize=(15,6))
sns.barplot(x='Category', y='Installs', data=category_installs)
plt.xticks(rotation=90)
plt.title('Total Installs by Category')
plt.show()
```

```python
# Top 5 apps by installs in each category
top5_apps = (
    df.sort_values(['Category', 'Installs'], ascending=[True, False])
      .groupby('Category')
      .head(5)
)

print(top5_apps[['Category', 'App', 'Installs']])


```

```python
plt.figure(figsize=(15,8))
sns.barplot(
    data=top5_apps,
    x='App',
    y='Installs',
    hue='Category'
)

plt.xticks(rotation=90)
plt.title('Top 5 Installed Apps in Each Category')
plt.show()
```

```python
rating_counts = df['Rating'].value_counts().sort_index()

plt.figure(figsize=(10,5))
sns.barplot(x=rating_counts.index, y=rating_counts.values)
plt.xlabel("Rating")
plt.ylabel("Number of Apps")
plt.title("Number of Apps by Rating")
plt.show()
```
